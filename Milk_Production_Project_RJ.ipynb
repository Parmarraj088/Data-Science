```python
import pandas as pd

# Load the dataset
df = pd.read_csv('monthly-milk-production-pounds-p.csv')

# Display the first few rows of the DataFrame
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Plot the time series
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Monthly milk production: pounds per cow. Jan 62 ? Dec 75'])
plt.xlabel('Month')
plt.ylabel('Milk Production (Pounds per Cow)')
plt.title('Monthly Milk Production')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# ADF test for stationarity
result = adfuller(df['Monthly milk production: pounds per cow. Jan 62 ? Dec 75'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Build ARIMA model
model = ARIMA(df['Monthly milk production: pounds per cow. Jan 62 ? Dec 75'], order=(1,1,1))
model_fit = model.fit()

# Summary of the model
print(model_fit.summary())
```

```python
# Forecast
forecast = model_fit.forecast(steps=12)  # Example forecast for 12 months
print('Forecast:', forecast)
```
